```python
import pandas as pd
import datetime
import numpy as np
from pybaseball import statcast_pitcher
from pybaseball import playerid_lookup
```

```python
today = datetime.datetime.today().strftime('%Y-%m-%d')
```

```python
def getPercents(ln,fn):
    if (fn == 'Luis'):
        ide = 622491
    else:
        ide = playerid_lookup(ln,fn)['key_mlbam'][0]
    data = statcast_pitcher('2019-03-28', today, player_id = ide)
    cleaned = data[['pitch_type','balls','strikes']]    
    typePitches = cleaned.groupby(['pitch_type','balls', 'strikes']).size().reset_index().rename(columns={0:'count'})
    totalPitches = cleaned.groupby(['balls', 'strikes']).size().reset_index().rename(columns={0:'count'})
    df3 = pd.merge(typePitches, totalPitches, on=["balls", "strikes"], how="right")
    df3['avg'] = df3['count_x'] / df3['count_y']
    new = df3[['pitch_type','balls','strikes','avg']]
    string = 'data/' + ln + '_' + fn + '.csv'
    new.to_csv(string)
```

```python
getPercents('Ryu','Hyun-Jin')
getPercents('Castillo','Luis')
getPercents('Scherzer','Max')
getPercents('Soroka','Mike')
getPercents('Kershaw','Clayton')
getPercents('Buehler','Walker')
```
