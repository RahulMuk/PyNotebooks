```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup, Comment
import json
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
totalData = []
url = "https://baseballsavant.mlb.com/player/full_spray?playerId=545361"
json_url = urlopen(url)
data = json.loads(json_url.read())
df = pd.DataFrame(data)
df
```

```python
df.to_csv('troutSpray.csv')
```

```python
df_hits = df[(df['events']!='force_out') & (df['events']!='field_out')]
```

```python
df_hits
```

```python
df_hits = df[(df['events']=='single') | (df['events']=='double') | (df['events']=='triple') | (df['events']=='home_run')]
df_hits["hc_x"] = pd.to_numeric(df_hits["hc_x"])
df_hits["hc_x_ft"] = pd.to_numeric(df_hits["hc_x_ft"])
df_hits["hc_y"] = pd.to_numeric(df_hits["hc_y"])
df_hits["hc_y_ft"] = pd.to_numeric(df_hits["hc_y_ft"])
df_avg = pd.pivot_table(df_hits,index=['year'], columns = 'events', values = ["hc_y","hc_x","hc_y_ft","hc_x_ft"])
```

```python
df_avg
```

```python
df_avg.to_csv('plots.csv')
```
